```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Electric_Production (1).csv')
```

```python
df.head()
```

```python
plt.scatter(df['DATE'],df['IPG2211A2N'],color='red')
```

```python
x1=df['IPG2211A2N'].ewm(alpha=0.5).mean()
```

```python
df['ewma']=x1
```

```python
plt.scatter(df['DATE'],df['IPG2211A2N'],color='red')
plt.plot(df['DATE'],x1,color='black',label='alpha=0.9')
plt.legend()
plt.show()
```
