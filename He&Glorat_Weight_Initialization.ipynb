```python
df=pd.read_csv('/content/placement (2).csv')
```

```python
df.head()
```

```python
x=df.iloc[:,0:2].values
y=df.iloc[:,-1].values
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

**Without Weight Initialization**

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y, clf=model, legend=2)
```

**He Initialization**

```python
model=Sequential()

model.add(Dense(10,activation='relu',input_shape=(2,),kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y,clf=model,legend=2)
```

**Glorat Initialization**

```python
model = Sequential()

model.add(Dense(10,activation='tanh',input_shape=(2,),kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y,clf=model,legend=2)
```
